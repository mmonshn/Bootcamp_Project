```python
# final exam 150 questions, pass >= 120
def grade(score):
    if score >= 120:
        return "Passed"
    elif score >= 100:
        return "Good"
    elif score >= 80:
        return "Okay"
    else:
        return "Need to read more!"
```

```python
result = grade(95)
print(result)
```

```python
# Use and, or in condition
# course == data science, score >= 80 Passed
# course == english, score >= 70 Passed

def grade(course, score):
    if course == "english" and score >= 70:
        return "Passed"
    elif course == "data science" and score >= 80:
        return "Passed"
    else:
        return "Failed"
```

```python
grade("data science", 81)
```

```python
# for loop
# if score >= 80, Passed
def grading_all(scores):
    new_scores = []
    for score in scores:
        new_scores.append(score-2)
    return new_scores

#scores = [88, 90, 75]
#new_scores = []
#for score in scores:
#   new_scores.append(score-2)

#print(new_scores)
```

```python
class Dog:
    def __init__(self, name, age, breed):
        self.name = name
        self.age = age
        self.breed = breed
```

```python
dog1 = Dog("Ovaltine", 2, "Chihuahua")
dog2 = Dog("Milo", 3, "Bulldog")
dog3 = Dog("Pepsi", 3.5, "German Shepherd")
```

```python
print(dog1.name, dog1.age, dog1.breed)
print(dog2.name, dog2.age, dog2.breed)
print(dog3.name, dog3.age, dog3.breed)
```

```python
# Object: attribute => name, id, dept, pos
# Object: method => hello(), work_hours(), chang_dept()
```

```python
class Employee:
    def __init__(self, id, name, dept, pos):
        self.id = id
        self.name = name
        self.dept = dept
        self.pos = pos # position

    def hello(self):
        print(f"Hello! my name is {self.name}")

    def work_hours(self, hours):
        print(f"{self.name} works for {hours} hours.")

    def change_dept(self, new_dept):
        self.dept = new_dept
        print(f"{self.name} is now in {self.dept}.")
```

```python
emp1 = Employee(1, "John", "Finance", "Financial Analyst")
```

```python
print(emp1.name, emp1.pos)
```

```python
emp1.hello()
```

```python
emp1.work_hours(10)
```

```python
emp1.dept
```

```python
emp1.change_dept("Data Science")
```
