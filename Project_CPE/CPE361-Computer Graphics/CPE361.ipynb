# **Line**

```python
import matplotlib.pyplot as plt

# function to plot the line using Bresenham's algorithm
def draw_line(x1, y1, x2, y2):
    # calculate the differences between the two points
    dx = x2 - x1
    dy = y2 - y1

    # determine whether to increment x or y
    if dx < 0:
        stepx = -1
        dx = -dx
    else:
        stepx = 1

    if dy < 0:
        stepy = -1
        dy = -dy
    else:
        stepy = 1

    # calculate the initial error
    error = dx - dy

    # plot the initial point
    plt.plot(x1, y1, 'bo')

    # initialize the current point
    x = x1
    y = y1

    # loop through each pixel along the line
    while x != x2 or y != y2:
        # calculate the error for the next pixel
        e2 = 2 * error

        if e2 > -dy:
            error = error - dy
            x = x + stepx

        if e2 <= dx:
            error = error + dx
            y = y + stepy

        # plot the current point
        plt.plot(x, y, 'bo')

        # print the current point and error
        print(f"({x}, {y}), error = {error}")

    # show the plot
    plt.show()

# get the starting and ending points from the user
x1 = int(input("Enter the x-coordinate of the starting point: "))
y1 = int(input("Enter the y-coordinate of the starting point: "))
x2 = int(input("Enter the x-coordinate of the ending point: "))
y2 = int(input("Enter the y-coordinate of the ending point: "))

# draw the line using Bresenham's algorithm
draw_line(x1, y1, x2, y2)
```

# **Circle**

```python
def bresenham_circle(x, y, r):
    points = set()
    d = 3 - 2 * r
    x_coord = 0
    y_coord = r
    
    while x_coord <= y_coord:
        points.add((x + x_coord, y + y_coord))
        points.add((x + x_coord, y - y_coord))
        points.add((x - x_coord, y + y_coord))
        points.add((x - x_coord, y - y_coord))
        points.add((x + y_coord, y + x_coord))
        points.add((x + y_coord, y - x_coord))
        points.add((x - y_coord, y + x_coord))
        points.add((x - y_coord, y - x_coord))
        
        if d < 0:
            d += 4 * x_coord + 6
        else:
            d += 4 * (x_coord - y_coord) + 10
            y_coord -= 1
        
        x_coord += 1
    
    sorted_points = sorted(points, key=lambda p: (p[0], -p[1]))
    return sorted_points

center_x, center_y, radius = 5, 5, 5
points = bresenham_circle(center_x, center_y, radius)

for point in points:
    print(point)
```

```python
def midPointCircleDraw(x_centre, y_centre, r):
    x = r
    y = 0
    points = set()

    # Adding the initial point at the top of the circle
    points.add((x + x_centre, y + y_centre))

    # When radius is zero, only a single point will be added
    if r > 0:
        points.add((x + x_centre, -y + y_centre))

    # Initializing the value of P
    P = 1 - r

    while x > y:
        y += 1

        # Mid-point inside or on the perimeter
        if P <= 0:
            P = P + 2 * y + 1

        # Mid-point outside the perimeter
        else:
            x -= 1
            P = P + 2 * y - 2 * x + 1

        # Adding the generated point and its reflection
        # in the other octants after translation
        if x >= y:
            points.add((x + x_centre, y + y_centre))
            points.add((-x + x_centre, y + y_centre))
            points.add((x + x_centre, -y + y_centre))
            points.add((-x + x_centre, -y + y_centre))

        # If the generated point lies on the line x = y,
        # only add one of the four symmetric points
        if x != y:
            points.add((y + x_centre, x + y_centre))
            points.add((y + x_centre, -x + y_centre))
            points.add((-y + x_centre, x + y_centre))
            points.add((-y + x_centre, -x + y_centre))

    # Sort the points based on their quadrant
    sorted_points = sorted(points, key=lambda p: (-p[0] + x_centre, p[1] - y_centre))

    # Printing the sorted points
    for point in sorted_points:
        print("(", point[0], ", ", point[1], ")", sep="", end="")

# Driver Code
if __name__ == '__main__':
    # To draw a circle of radius 5
    # centered at (5, 5)
    midPointCircleDraw(5, 5, 5)
```

# **Parallel Projections**

```python
import csv
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

filename = "OBJECT1.csv"
values = []

# Open the file and read its contents
with open(filename, 'r') as csvfile:
    csvreader = csv.reader(csvfile)

    # Loop through each row in the file
    for row in csvreader:
        # Loop through each column in the row
        for value in row:
            # Add the value to the list
            values.append(value)
values = [values for values in values if values != '']
# Print out the values
print(values)

# Create a list of polygons
polygons = []
i = 0

while i < len(values):
    if values[i] == 'POLY':
        num_vertices = int(values[i+1])
        # read the next `num_vertices` groups of three numbers as coordinates
        polygon_vertices = []
        for j in range(num_vertices):
            x = float(values[i+j*3+2])
            y = float(values[i+j*3+3])
            z = float(values[i+j*3+4])
            polygon_vertices.append((x, y, z))
        polygons.append(polygon_vertices)
        i += num_vertices * 3 + 2 # skip to the next "POLY"
    else:
        i += 1

# Plot the polygons
fig = plt.figure(figsize=(10, 10))

# Top View
ax1 = fig.add_subplot(221)
for polygon in polygons:
    # add the first vertex to the end of the list to close the polygon
    polygon.append(polygon[0])
    # convert the polygon vertices to separate lists of x, y, and z coordinates
    x, y, z = zip(*polygon)
    # plot the polygon as a line between the vertices
    ax1.plot(x, y)
ax1.set_aspect('equal')
ax1.set_xlim(min(x), max(x))
ax1.set_ylim(min(y), max(y))
ax1.set_title('Top View')

# Front View
ax2 = fig.add_subplot(222, projection='3d')
for polygon in polygons:
    # add the first vertex to the end of the list to close the polygon
    polygon.append(polygon[0])
    # convert the polygon vertices to separate lists of x, y, and z coordinates
    x, y, z = zip(*polygon)
    # plot the polygon as a line between the vertices
    ax2.plot(x, y, z)
ax2.view_init(elev=0, azim=0)
ax2.set_title('Front View')

# Side View
ax3 = fig.add_subplot(223, projection='3d')
for polygon in polygons:
    # add the first vertex to the end of the list to close the polygon
    polygon.append(polygon[0])
    # convert the polygon vertices to separate lists of x, y, and z coordinates
    x, y, z = zip(*polygon)
    # plot the polygon as a line between the vertices
    ax3.plot(x, y, z)
ax3.view_init(elev=0, azim=90)
ax3.set_title('Side View')

plt.show()
```

# **Hermite Curve**

```python
import numpy as np
import matplotlib.pyplot as plt

def hermite_curve(start_point, end_point, start_tangent, end_tangent, num_points):
    t = np.linspace(0, 1, num_points)
    h00 = 2 * t**3 - 3 * t**2 + 1
    h10 = t**3 - 2 * t**2 + t
    h01 = -2 * t**3 + 3 * t**2
    h11 = t**3 - t**2

    x = start_point[0] * h00 + end_point[0] * h01 + start_tangent[0] * h10 + end_tangent[0] * h11
    y = start_point[1] * h00 + end_point[1] * h01 + start_tangent[1] * h10 + end_tangent[1] * h11

    return x, y

# Example usage
start_point = (-4, -1)
end_point = (-4, 4)
start_tangent = (3, 1)
end_tangent = (1, 1)
num_points = 100

x, y = hermite_curve(start_point, end_point, start_tangent, end_tangent, num_points)

# Plotting the curve
plt.plot(x, y)
plt.scatter([start_point[0], end_point[0]], [start_point[1], end_point[1]], color='red')
plt.quiver(start_point[0], start_point[1], start_tangent[0], start_tangent[1], color='blue')
plt.quiver(end_point[0], end_point[1], end_tangent[0], end_tangent[1], color='blue')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Hermite Curve')
plt.grid(True)
plt.show()
```

# **Bezier Curve BezierCastel**

```python
import matplotlib.pyplot as plt
import numpy as np

def de_casteljau(control_points, t):
    if len(control_points) == 1:
        return control_points[0]
    else:
        new_points = []
        for i in range(len(control_points) - 1):
            x = (1 - t) * control_points[i][0] + t * control_points[i+1][0]
            y = (1 - t) * control_points[i][1] + t * control_points[i+1][1]
            new_points.append((x, y))
        return de_casteljau(new_points, t)

control_points = [(0, 20), (10, 40), (20, 0), (30, 20)]
t_values = np.linspace(0, 1, 101)
curve_points = [de_casteljau(control_points, t) for t in t_values]

x = [point[0] for point in curve_points]
y = [point[1] for point in curve_points]

fig, ax = plt.subplots()
ax.plot(x, y)
ax.plot([point[0] for point in control_points], [point[1] for point in control_points], 'ro')
ax.set_aspect('equal')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```

```python
import matplotlib.pyplot as plt

def de_casteljau(t, p0, p1, p2, p3):
    """Perform de Casteljau algorithm to find point on Bezier curve"""
    q0 = (1-t)*p0 + t*p1
    q1 = (1-t)*p1 + t*p2
    q2 = (1-t)*p2 + t*p3
    r0 = (1-t)*q0 + t*q1
    r1 = (1-t)*q1 + t*q2
    return (1-t)*r0 + t*r1

# Get user input for control points
x0, y0 = map(float, input("Enter x0 and y0 separated by a space: ").split())
x1, y1 = map(float, input("Enter x1 and y1 separated by a space: ").split())
x2, y2 = map(float, input("Enter x2 and y2 separated by a space: ").split())
x3, y3 = map(float, input("Enter x3 and y3 separated by a space: ").split())

# Generate points on Bezier curve using de Casteljau algorithm
points = []
for i in range(101):
    t = i/100
    x = de_casteljau(t, x0, x1, x2, x3)
    y = de_casteljau(t, y0, y1, y2, y3)
    points.append((x,y))

# Plot control points and Bezier curve
plt.plot([x0,x1,x2,x3], [y0,y1,y2,y3], 'ro')
plt.plot([p[0] for p in points], [p[1] for p in points])
plt.show()
```
